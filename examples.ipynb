```python
import cv2
from matplotlib import pyplot as plt
from skimage import color
from metrics import AMBE, IEM,UCIQE,UIQM

imageA = cv2.imread('images/Fig0310(b)(washed_out_pollen_image).tif',0)
imageB = cv2.equalizeHist(imageA)

plt.subplot(121)
plt.imshow(imageA, vmin=0, vmax=255, cmap='gray')
plt.subplot(122)
plt.imshow(imageB, vmin=0, vmax=255, cmap='gray')
plt.show()
```

# Absolute Mean Brightness Error (AMBE)

```python
ambe = AMBE(imageA, imageB)
print(f'AMBE is: {ambe}')
```

# Image Enhancement Metric(IEM)

```python
iem = IEM(imageA, imageB)
print(f'IEM is: {iem}')
```

# Underwater colour image quality evaluation metric (UCIQE)

```python
color_imageA_GT=cv2.imread('images/01_outdoor_GT.jpg') # cv2.imread() ->formato BGR por padro
color_imageA_GT=cv2.resize(color_imageA_GT,(360,240))
color_imageA_hazy=cv2.imread('images/01_outdoor_hazy.jpg')
color_imageA_hazy=cv2.resize(color_imageA_hazy,(360,240))

# converso do espao de cores
color_imageA_GT=cv2.cvtColor(color_imageA_GT,cv2.COLOR_BGR2RGB)
color_image_hazy=cv2.cvtColor(color_imageA_hazy,cv2.COLOR_BGR2RGB)

# calculo
uciqe_color_imageA_GT=UCIQE(color_imageA_GT)
uciqe_color_imageA_hazy=UCIQE(color_imageA_hazy)

# plot
fig = plt.figure(figsize=(9,6))  
plt.subplot(121); plt.axis("off"); plt.imshow(color_imageA_GT);plt.subplot(121).set_title(f"UCIQUE: {uciqe_color_imageA_GT}",)
plt.subplot(122); plt.axis("off"); plt.imshow(color_imageA_hazy);plt.subplot(122).set_title(f"UCIQUE: {uciqe_color_imageA_hazy}")
plt.show()
```

# Underwater image quality measure (UIQM)

```python
# calculo
uiqm_color_imageA_GT=UIQM(color_imageA_GT)
uiqm_color_imageA_hazy=UIQM(color_imageA_hazy)

# plot
fig = plt.figure(figsize=(9,6))  
plt.subplot(121); plt.axis("off"); plt.imshow(color_imageA_GT);plt.subplot(121).set_title(f"UIQM: {uiqm_color_imageA_GT}",)
plt.subplot(122); plt.axis("off"); plt.imshow(color_imageA_hazy);plt.subplot(122).set_title(f"UIQM: {uiqm_color_imageA_hazy}")
plt.show()
```
