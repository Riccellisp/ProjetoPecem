```python
db = pd.read_csv('db_pecem.csv')

for cam in cam_paths:
    cam_num = cam.split('/')[-1]
    print(cam_num)
    ind_cam_path = [os.path.join(cam,x) for x in os.listdir(cam)]
    
    for ind_cam in ind_cam_path:
        image_path = ind_cam
        ind_cam = ind_cam.split('/')[-1]
        image_name = ind_cam
        
        # cameras with xh in name
        if 'h' in ind_cam:
            get_hour = ind_cam.split(' - ')[-1]
            get_hour = get_hour.replace('.jpg', '')
            get_hour = re.findall(r'\d+', get_hour)
            if int(get_hour[0]) < 18 and int(get_hour[0]) > 6:
                shift = 'dia'
            else:
                shift = 'noite'
                
        elif 'Imagem' in ind_cam:
            shift = 'dia'
                
        else:
            get_hour = ind_cam.split(' - ')[-1]
            get_hour = re.findall(r'\d+', get_hour)
            if int(get_hour[3]) < 18 and int(get_hour[3]) > 6:
                shift = 'dia'
            else:
                shift = 'noite'

        new_row = {'image_name': image_name, 'image_path': image_path,
           'cam_num': cam_num, 'shift': shift,
           'ref_excelente':dict_ref[cam_num][shift+'_Excelente'],
           'ref_bom':dict_ref[cam_num][shift+'_Bom'],
           'ref_ruim':dict_ref[cam_num][shift+'_Ruim'],
           'ref_pessimo':dict_ref[cam_num][shift+'_Pessimo'],
          
          
          }

        db = db.append(new_row, ignore_index=True)
        
                
        
```
